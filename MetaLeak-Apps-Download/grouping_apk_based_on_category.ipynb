```python
def checkFileOrDirectoryExist(path):
    if os.path.exists(path):
        return True
    else:
        return False
```

```python
def createDirectory(path):
    os.mkdir(path)
```

```python
def copyApkFile(source,destination):
    shutil.copyfile(source, destination)
```

```python
# 1. Read category apk file
# 1.1 Copy list file
source_csv = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\grouping_apk.csv"
dest_csv = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\category_apk\\grouping_apk.csv"
copyApkFile(source_csv,dest_csv)
# 1.2 Read file
# df=""
# if(checkFileOrDirectoryExist("C:\\Users\\ASUS\\anaconda3\\phd_implement\\category_apk\\grouping_apk_done.csv")):
#     print("Read file grouping_apk_done.csv ")
#     df = pd.read_csv("C:\\Users\\ASUS\\anaconda3\\phd_implement\\category_apk\\grouping_apk_done.csv",low_memory=False)
# else:
#     print("No file grouping_apk_done.csv - Read file grouping_apk.csv")
#     df = pd.read_csv("C:\\Users\\ASUS\\anaconda3\\phd_implement\\category_apk\\grouping_apk.csv",low_memory=False)
df = pd.read_csv("C:\\Users\\ASUS\\anaconda3\\phd_implement\\category_apk\\grouping_apk.csv",low_memory=False)
df.head()
```

```python
# 2. Create directory for apps with directory's name = apk_category
# 2.1 Get dowloaded index
index_downloaded = df[(df.download !="Done") & (df.download !="Not Found")].index[0]
#index_downloaded
#apk_category = df[["apk_category","apkfile_name"]][0:index_downloaded]
# 2.2 Get classified index
classify_index = df[(df.classify =="Done") | (df.classify =="Remove-Done")].index[-1]
apk_category = df["apk_category"][classify_index:index_downloaded]
df_apk_category = pd.DataFrame(apk_category)
print("index_downloaded= "+str(index_downloaded))
print("classify_index= "+str(classify_index))
print(df_apk_category)
```

```python
category_group = df_apk_category.groupby('apk_category')
category_list_amount = category_group.size().values.tolist()
category_list_name = category_group.nunique().index.values.tolist()
print(category_list_name)
print(len(category_list_name))
print(category_list_amount)
print(len(category_list_amount))
```

```python
directory_path = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\category_apk\\"
for i in range(len(category_list_name)):
    if not checkFileOrDirectoryExist(directory_path+category_list_name[i]):
        createDirectory(directory_path+category_list_name[i])
```

```python
# 3.Classified APK file to folder
# 3.1 Get data
apk_category = df[["apk_category","apkfile_name"]][classify_index:index_downloaded]
print(apk_category)  
```

```python
#apkfile_name_group = apk_category.groupby(['apk_category', 'apkfile_name'])
#apkfile_name_group.size()
```

```python
# 3.2 Classify APK
grouped_df = apk_category.groupby("apk_category")
dict_category = {}
for key, item in grouped_df:
    #print(grouped_df.get_group(key))
    apkfile_name_group = grouped_df.get_group(key)
    #print(apkfile_name_group["apkfile_name"])
    #print(apkfile_name_group["apkfile_name"][98])
    array_category = []
    for j in range(len(apkfile_name_group)):
        #print(apkfile_name_group.index[j])
        array_category.append(apkfile_name_group["apkfile_name"][apkfile_name_group.index[j]])
    dict_category[key]=array_category
print(len(dict_category))
```

```python
# 3.4 Move apk file to folder
source_path = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\download_apk\\"
category_path = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\category_apk\\"
for x in range(len(category_list_name)):
    category_name = str(category_list_name[x])
    print("------------------"+category_name+"------------------")
    #print(category_name)
    arr_apk = dict_category[category_name]
    #print(arr_apk)
    for y in range(len(arr_apk)):
        #print(arr_apk[y])
        print("["+str(y)+"]-----------------------------------------------")
        source = source_path+str(arr_apk[y])
        destination = category_path+category_name+"\\"+str(arr_apk[y])
        print(source)
        print(destination)
        print(checkFileOrDirectoryExist(source))
        print(checkFileOrDirectoryExist(destination))
        if (checkFileOrDirectoryExist(source)):
            if (not checkFileOrDirectoryExist(destination)):
                copyApkFile(source,destination)
                print("Moved "+str(arr_apk[y]))
                df.loc[df["apkfile_name"] == arr_apk[y], "classify"] = "Done"
            else:
                print("APK "+str(arr_apk[y])+" classified")
                df.loc[df["apkfile_name"] == arr_apk[y], "classify"] = "Done"
        else:
                print("APK "+str(arr_apk[y])+" removed")
        #break
    #break
```
