```python
def waitingDownload(category_name: str):
    path = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\tmp\\*.apk"
    path_xapk = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\tmp\\*.xapk"
    files = glob.glob(path)
    files_xapk = glob.glob(path_xapk)
    #print(files)
    if(len(files)!=0):
        return True
    elif(len(files_xapk)!=0):
        return True
    else:
        return False
```

```python
def deleteFile(dir):
    for f in os.listdir(dir):
        os.remove(os.path.join(dir, f))
```

```python
def removeInvalidCharacter(filename):
    invalid = ' <>:"/\|?*!+;,=()'
    for char in invalid:
        filename = filename.replace(char,'-')
    return filename
```

```python
def fileAPKPathFunc(category_name:str):
    path = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\tmp\\*.apk"
    path_xapk = "C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\tmp\\*.xapk"
    files = glob.glob(path)
    files_xapk = glob.glob(path_xapk)
    #print(files)
    if((len(files)!=0)):
        return files[0],"apk"
    elif((len(files_xapk)!=0)):
        return files_xapk[0],"xapk"
    else:
        return "",""
```

```python
def checkFileOrDirectoryExist(path):
    if os.path.exists(path):
        return True
    else:
        return False
```

```python
def writeNextCSVIndex(path,index):
    with open(path, 'w') as f:
        f.write(index)
```

```python
def readNextCSVIndex(path):
    f = open(path, "r")
    return int(f.read())
```

```python
# 0. 
# 0.1 Read file
category_name = "dating"
apk_zone = "top-new-free"
df = pd.read_csv("C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\"+category_name+"_apk_popular.csv",low_memory=False)
# 0.2 Determine continue download
index_downloaded = 0
current_download_pages = 1
current_download_link = ""
if(df.size!=0):
    index_downloaded = df[(df.download =="Done")].index[-1]
    current_download_pages = int(df.page[index_downloaded])
    current_download_link = str(df.apk_link[index_downloaded])
else:
    index_downloaded = 0
    current_download_pages = 1
    current_download_link = ""
# LAM FIRST TRIGGER
#current_download_pages = 1
#current_download_link = ""
#LAM FIRST TRIGGER
print("Current Download Pages: "+str(current_download_pages))
print("Current Download Links: "+current_download_link)
print("Current Index Downloaded: "+str(index_downloaded))
```

```python
try:
    # 1.Get list app download link
    # 1.1 Setup selenium
    chromedriver = 'chromedriver.exe'
    os.environ["webdriver.chrome.driver"] = chromedriver
    chrome_options = Options()
    chrome_options.add_argument("--disable-blink-features=AutomationControlled")
    chrome_options.add_argument("--disable-popup-blocking")
    chrome_options.add_argument('--ignore-ssl-errors=yes')
    chrome_options.add_argument('--ignore-certificate-errors')
    chrome_options.add_argument('--allow-running-insecure-content')
    chrome_options.add_argument('--no-proxy-server')
    # open web in background
    chrome_options.add_argument('--headless=new')
    # open web in background
    prefs = {
        'profile.default_content_setting_values.automatic_downloads': 1,
        "download.default_directory" : "C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\tmp"
    }
    chrome_options.add_experimental_option("prefs", prefs)
    # 1.2 Open Web Page
    num_page = 101
    for num in range(int(current_download_pages),num_page):
        continue_download_index=0
        web = webdriver.Chrome(options=chrome_options)
        web.maximize_window()
        apk_path = "https://apkcombo.com/category/"+category_name+"/"+apk_zone+"/?page="+str(num)
        print(apk_path)
        web.get(apk_path)
        time.sleep(10)
        print("----------------------------------[Page-"+str(num)+"]----------------------------------")
        # 1.3 Agree privacy
        try:
            #privacy_agree=web.find_element(By.CSS_SELECTOR,"button[class='sc-ifAKCX ljEJIv']")
            privacy_agree=web.find_element(By.CSS_SELECTOR,"button[class='fc-button fc-cta-consent fc-primary-button']")
            privacy_agree.click()
            time.sleep(5)
        except NoSuchElementException:
            print("Element does not exist")
        #time.sleep(5)
        # 1.4 Get list apk
        apk_lists=web.find_elements(By.CSS_SELECTOR,"a[class='l_item']")
        #print(len(apk_lists))
        download_link_array = []
        for i in range(len(apk_lists)):
            pkg_name=web.find_elements(By.CSS_SELECTOR,"a[class='l_item']")[i].get_attribute('href')
            download_link_array.append(pkg_name+"download/apk")
        print(download_link_array)
        if(current_download_link in download_link_array):
            continue_download_index = download_link_array.index(current_download_link)
        else:
            print("This is new page")
            continue_download_index = -1
        print("Continue download index "+str(continue_download_index+1))
        continue_download_arr=download_link_array[continue_download_index+1:]
        print(continue_download_arr)
        print(len(continue_download_arr))
        if(len(continue_download_arr)==0):
            current_download_pages = int(current_download_pages) + 1
            web.close()
            continue
        web.close()
        # 2. Download apk
        for j in range(len(continue_download_arr)):
            #csv_index = str(int(str(num)+str(j)) +continue_download_index)
            csv_index = readNextCSVIndex("C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\index.txt") 
            print("csv_index= "+str(csv_index))
            # 2.1 Open web page
            web = webdriver.Chrome(options=chrome_options)
            web.maximize_window()
            web.get(continue_download_arr[j])
            time.sleep(7)
            # 2.2 Agree privacy
            try:
                #privacy_agree=web.find_element(By.CSS_SELECTOR,"button[class='sc-ifAKCX ljEJIv']")
                privacy_agree=web.find_element(By.CSS_SELECTOR,"button[class='fc-button fc-cta-consent fc-primary-button']")
                privacy_agree.click()
                time.sleep(5)
                #df.loc[int(csv_index),"apk_link"]=str(continue_download_arr[j])
            except NoSuchElementException:
                print("Element does not exist")
            # 2.3 Find download link 
            try:
                apk_link=web.find_element(By.CLASS_NAME, "variant").get_attribute('href')
                apkfile_name = web.find_element(By.CLASS_NAME, "vername").text
                apkfile_type = web.find_element(By.CLASS_NAME, "type-apk").text
                print("[Link-"+str(j)+"] "+apk_link)
                apkfile_name=removeInvalidCharacter(apkfile_name)
                print("File name: "+apkfile_name)
                pkg_name_arr = (continue_download_arr[j]).split("/")
                pkg_name = str(pkg_name_arr[4])
                print("Package ID: "+pkg_name)
                df.loc[int(csv_index),"pkg_name"]=pkg_name
                if(apkfile_type!=""):
                    df.loc[int(csv_index),"apkfile_name"]=str(apkfile_name+"."+(apkfile_type).lower())
                else:
                    df.loc[int(csv_index),"apkfile_name"]=str(apkfile_name+".xapk")
            except NoSuchElementException:
                print("[Link-"+str(j)+"] App does not exist")
                continue
            # 2.2 Download APK  files from download_link
            def downloadAPK():
                web.get(apk_link)
            thread= Thread(target=downloadAPK())
            thread.start()
            print('Waiting for the apk downloading...')
            print('Start timeout...')
            timeout = time.time() + 60*15 # timeout after 15 minutes
            while True:
                if((waitingDownload(category_name)) or (time.time() > timeout)):
                    break
            thread.join()
            print("apk_link: "+str(continue_download_arr[j]))
            df.loc[int(csv_index),"apk_link"]=str(continue_download_arr[j])
            print('Download completed')
            time.sleep(7)
            df.loc[int(csv_index),"download"]="Done"
            df.loc[int(csv_index),"page"]=int(num)
            time.sleep(7)
            fileAPKPath,fileExtension=fileAPKPathFunc(category_name)
            if((fileAPKPath!="") and (fileExtension!="")):
                if(not checkFileOrDirectoryExist("C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\"+str(apkfile_name)+"."+str(fileExtension))):
                    time.sleep(15)
                    os.rename(fileAPKPath, "C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\"+str(apkfile_name)+"."+str(fileExtension))
                    time.sleep(15)
                    deleteFile("C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\tmp\\")
                    time.sleep(5)
                else:
                    deleteFile("C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\tmp\\")
                    time.sleep(10)
            writeNextCSVIndex("C:\\Users\\ASUS\\anaconda3\\phd_implement\\"+category_name+"_category\\index.txt",str(csv_index+1))
            web.close()
            time.sleep(7)
except KeyboardInterrupt:
    exportResultCSV(df)
except:
    exportResultCSV(df)
```
